```python
# otimizando o código de tratamento do texto

# importando a função "regex" para tokenização
from nltk.tokenize import RegexpTokenizer

# instanciando o caminho do arquivo de texto
corpus = open('/content/Os 10 melhores poemas de Fernando Pessoa.txt').read()

# instanciando a função "regex" na variável "tokenizer"
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') # ignorando todos os símbolos e números

# aplicando a tokenização e salvando o resultado na variável "tokens"
tokens = tokenizer.tokenize(corpus)

# instanciando a lista de stopwords na variável "stopwords"
stopwords = nltk.corpus.stopwords.words('portuguese')

# criando uma list comprehesion para tratar o arquivo de texto
texto_tratado = [token.lower() for token in tokens if token.lower() not in stopwords]
```

```python
texto_tratado
```

```python
# imprimindo as frequências das palavras
freq = nltk.FreqDist(texto_tratado)
freq.most_common(30)  # visualizando as 30 palavras mais frequentes
```

```python
# visualizando as 30 palavras mais frequentes - unigramas
nlp_words = nltk.FreqDist(texto_tratado)
plot1 = nlp_words.plot(30, color = 'salmon', 
                       title = 'Frequência de palavras - unigramas')
```

```python
# importando as funções "bigrams" e "trigrams"
from nltk import bigrams, trigrams
```

```python
tokens = texto_tratado
tokens_bigrams = list(bigrams(tokens))

tokens_bigrams[:10]
```

```python
# visualizando as 30 palavras mais frequentes - bigramas
bigrm = list(nltk.bigrams(texto_tratado))
words2 = nltk.FreqDist(bigrm)
words2.plot(30, color = 'salmon', 
           title = 'Frequência de palavras - bigramas')
```

```python
tokens = texto_tratado
tokens_trigrams = list(trigrams(tokens))

tokens_trigrams[:10]
```

```python
# visualizando as 30 palavras mais frequentes - trigramas
trigrm = list(nltk.trigrams(texto_tratado))
words3 = nltk.FreqDist(trigrm)
words3.plot(30, color = 'salmon', 
           title = 'Frequência de palavras - trigramas')
```

```python
# importando as bibliotecas necessárias
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# exemplos de comentários da lista "all_words"
display(texto_tratado[27:45])
display(texto_tratado[99])
display(texto_tratado[243])
display(texto_tratado[802])
```

```python
# realizando a concatenação das palavras
all_summary = ' '.join(s for s in texto_tratado)
```

```python
# atualizando a lista de stopwords
stopwords = set(STOPWORDS)
stopwords.update(['in', 'the', 'way', 'week', 'weeks', 'q', 'j', 't', 'est', 'batman', 'ago', 'deu'])
```

```python
# instanciando a lista de stopwords na variável "stopwords"
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
# gerando uma wordcloud - nuvem de palavras
wordcloud = WordCloud(stopwords=stopwords,
                      background_color='black',
                      width=1600, height=800, 
                      max_words=300).generate(all_summary)
 
# mostrando a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);

# salvando a imagem
wordcloud.to_file('poemasfernandopessoa_summary_wordcloud.png');
```
